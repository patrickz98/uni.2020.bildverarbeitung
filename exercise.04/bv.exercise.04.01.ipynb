```python
import math
import skimage.io
import numpy as np
import matplotlib.pyplot as plt

img1 = np.array(skimage.io.imread("bild1.png"), np.uint8)
newImg = np.zeros(img1.shape)

for inx in range(img1.shape[0]):
    for iny in range(img1.shape[1]):
        newImg[ inx ][ iny ] = 40 * math.log(1.0 + img1[ inx ][ iny ])

plt.figure()
plt.imshow(img1, cmap='gray', vmin=0, vmax=255)

plt.figure()
plt.imshow(newImg, cmap='gray', vmin=0, vmax=255)
skimage.io.imsave("bild1.new.png", newImg)
```

```python
img2 = np.array(skimage.io.imread("bild2.png"), np.uint8)

newImg = np.zeros(img2.shape)

for inx in range(img2.shape[0]):
    for iny in range(img2.shape[1]):
        newImg[ inx ][ iny ] = 0.00005 * img2[ inx ][ iny ] ** 3.0

plt.figure()
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)

plt.figure()
plt.imshow(newImg, cmap='gray', vmin=0, vmax=255)
skimage.io.imsave("bild2.new.png", newImg)
```

```python
img3 = np.array(skimage.io.imread("bild3.png"), np.uint8)

newImg = np.copy(img3)

for inx in range(img3.shape[0]):
    for iny in range(img3.shape[1]):
        if img3[ inx ][ iny ][ 0 ] > 100:
            newImg[ inx ][ iny ] = 0xff

plt.figure()
plt.imshow(img3, cmap='gray', vmin=0, vmax=255)

plt.figure()
plt.imshow(newImg, cmap='gray', vmin=0, vmax=255)
skimage.io.imsave("bild3.new.png", newImg)
```
